## JOB RECOMMENDATION

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

data_Recruiter = pd.read_csv("./job_post.csv")
data_Recruiter
```

```python
data_Candidate = pd.read_csv("./candidate_data.csv")
data_Candidate
```

```python
one_Candidate = data_Candidate.loc[data_Candidate['Name'] == 'Raj']
one_Candidate
```

```python
matched_jobs = pd.DataFrame()
for index,row in data_Recruiter.iterrows():
    for index2,row2 in one_Candidate.iterrows():
        if row2['Experience'] >= row['Experience'] and row2['Engineering'] >= row['Engineering'] and row2['10th'] >= row['10th'] and row2['12th'] >= row['12th']:
            matched_jobs = pd.concat([matched_jobs,pd.DataFrame(row).T],ignore_index=True)
matched_jobs
```

## Recruiter

```python
matched_jobs['text'] = matched_jobs['Position'].map(str) + " " +matched_jobs['Skills']
matched_jobs
```

```python
df_R = matched_jobs[['Company Name', 'text']]
df_R = df_R.fillna(" ")
df_R
```

## Function to Clean text

```python
from nltk.corpus import stopwords
import re
import string
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
stop = stopwords.words('english')
stop_words_ = set(stopwords.words('english'))
wn = WordNetLemmatizer()

def black_txt(token):
    return  token not in stop_words_ and token not in list(string.punctuation)  and len(token)>2   
  
def clean_txt(text):
  clean_text = []
  clean_text2 = []
  text = re.sub("'", "",text)
  text=re.sub("(\\d|\\W)+"," ",text) 
  text = text.replace("nbsp", "")
  clean_text = [ wn.lemmatize(word, pos="v") for word in word_tokenize(text.lower()) if black_txt(word)]
  clean_text2 = [word for word in clean_text if black_txt(word)]
  return " ".join(clean_text2)
```

```python
df_R['text'] = df_R['text'].apply(clean_txt)
df_R
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vectorizer = TfidfVectorizer()

tfidf_jobid = tfidf_vectorizer.fit_transform((df_R['text']))
tfidf_jobid
```

## Candidate

```python
one_Candidate['text'] = one_Candidate['Position'].map(str) + " " + one_Candidate['Skills']
one_Candidate
```

```python
df_C = one_Candidate[['Name', 'text']]
df_C = df_C.fillna(" ")
df_C
```

```python
df_C['text'] = df_C['text'].apply(clean_txt)
df_C
```

```python
from sklearn.metrics.pairwise import cosine_similarity
user_tfidf = tfidf_vectorizer.transform(df_C['text'])
cos_similarity_tfidf = map(lambda x: cosine_similarity(user_tfidf, x),tfidf_jobid)
```

```python
output2 = list(cos_similarity_tfidf)
```

```python
def get_recommendation(top, df_R, scores):
    recommendation = pd.DataFrame(columns = ['Company Name', 'score'])
    count = 0
    for i in top:
        recommendation.at[count, 'Company Name'] = df_R['Company Name'][i]
        recommendation.at[count, 'score'] =  scores[count]*100
        count += 1
    return recommendation
```

```python
top = sorted(range(len(output2)), key=lambda i: output2[i], reverse=True)[:10]
list_scores = [output2[i][0][0] for i in top]
recommendation = get_recommendation(top,df_R, list_scores)
recommendation
```
